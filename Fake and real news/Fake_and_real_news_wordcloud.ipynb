```python
import pandas as pd
from wordcloud import WordCloud ,STOPWORDS
import warnings
from collections import Counter
import matplotlib.pyplot as plt
warnings.filterwarnings("ignore")
```

```python
real = pd.read_csv("True.csv")
fake = pd.read_csv("Fake.csv", parse_dates=False)
```

### **Real data**

```python
real_text = ""
for txt in real.text:
    real_text += " "+txt
```

```python
# Create stopword list:
stopwords = set(STOPWORDS)

stopwords.update(["via", "even","many"])
# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords).generate(real_text)
```

```python
# Display the generated image:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Save the image in the img folder:
wordcloud.to_file("real_text.png")
```

```python
# Split the text into a list of individual words

tokens = real_text.split(" ")
```

```python
# Count the frequency of each word
word_counts = Counter(tokens)

# Display the count for a single word.
word_counts["trump"]
```

```python
# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords).generate_from_frequencies(word_counts)
```

```python
# Display the generated image:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
